```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Model parameters
alpha_A = 0.1
alpha_R = 0.1
beta = 1
K = 1
n = 2
delta_A = 0.1
delta_R = 0.1
delta_AR = 0.1

# ODE system for the model
def model(y, t):
    A, R, AR = y
    
    dA_dt = alpha_A - (beta * R * A) / (K + A) - delta_A * A
    dR_dt = alpha_R - delta_R * R
    dAR_dt = (beta * R * A) / (K + A) - delta_AR * AR
    
    return [dA_dt, dR_dt, dAR_dt]

# Initial conditions
A0 = 0
R0 = 5
AR0 = 0
y0 = [A0, R0, AR0]

# Time array
t = np.linspace(0, 100, 1000)

# Solve ODEs
sol = odeint(model, y0, t)
A, R, AR = sol.T

# Plotting results
fig, ax = plt.subplots(3, 1, figsize=(8, 12))

ax[0].plot(t, A, label="Autoinducer (A)")
ax[0].set_xlabel("Time")
ax[0].set_ylabel("A")
ax[0].legend()

ax[1].plot(t, R, label="Receptor (R)")
ax[1].set_xlabel("Time")
ax[1].set_ylabel("R")
ax[1].legend()

ax[2].plot(t, AR, label="Autoinducer-Receptor Complex (AR)")
ax[2].set_xlabel("Time")
ax[2].set_ylabel("AR")
ax[2].legend()

plt.show()
```
